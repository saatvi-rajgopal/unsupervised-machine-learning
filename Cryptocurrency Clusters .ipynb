```python
# Discard all the cryptocurrencies that are not being traded

crypto_df_prep = crypto_df.copy()
crypto_df_prep = crypto_df_prep.loc[crypto_df_prep.IsTrading == True].reset_index(drop=True)
crypto_df_prep.info()
```

```python
# Drop the 'IsTrading' column after applying filter

crypto_df_prep.drop(["IsTrading"], axis=1, inplace=True)
crypto_df_prep.head()
```

```python
# Remove all rows that have at least one null value

crypto_df_prep = crypto_df_prep.dropna(how="any").reset_index(drop=True)
crypto_df_prep.info()
```

```python
# Filter the cryptocurrencies that have been mined

crypto_df_prep = crypto_df_prep.loc[crypto_df_prep.TotalCoinsMined > 0].reset_index(drop=True)
crypto_df_prep.info()
```

```python
# Convert the 'TotalCoinSupply' column to numeric values

crypto_df_prep["TotalCoinSupply"] = pd.to_numeric(crypto_df_prep.TotalCoinSupply)
crypto_df_prep.info()
```

```python
# Look at the 'ProofType' list after the column drops

crypto_df_prep.ProofType.value_counts().index
```

```python
# Edit column values for the proper counts

crypto_df_prep.loc[crypto_df_prep.ProofType == "PoW and PoS", "ProofType"] = "PoW/PoS"
crypto_df_prep.loc[crypto_df_prep.ProofType == "Proof of Trust", "ProofType"] = "PoT"
crypto_df_prep.loc[crypto_df_prep.ProofType == "PoW/PoS ", "ProofType"] = "PoW/PoS"
crypto_df_prep.loc[crypto_df_prep.ProofType == "Pos", "ProofType"] = "PoS"
crypto_df_prep.loc[crypto_df_prep.ProofType == "Proof of Authority", "ProofType"] = "PoA"
```

```python
# Drop the columns that aren't providing any values

crypto_df_coin = crypto_df_prep.CoinName
crypto_df_prep.drop(["Unnamed: 0", "CoinName"], axis=1, inplace=True)
crypto_df_prep.head()
```

```python
# Convert the 'Algorithm' and 'ProofType' columns into numerical data

crypto_df_prep = pd.get_dummies(crypto_df_prep)
crypto_df_prep.head()
```

```python
# Standardize the 'TotalCoinsMined' and 'TotalCoinSupply' columns due to their large values

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler = scaler.fit(crypto_df_prep.loc[:, ["TotalCoinsMined", "TotalCoinSupply"]])

crypto_df_prep_scale = crypto_df_prep.copy()
crypto_df_prep_scale.loc[:, ["TotalCoinsMined", "TotalCoinSupply"]] = scaler.transform(crypto_df_prep.loc[:, ["TotalCoinsMined", "TotalCoinSupply"]])
crypto_df_prep_scale.head()
```

```python
# Cather the PCA explained variance
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(crypto_df_prep_scale)
print(np.sum(pca.explained_variance_ratio_))
print(pca.explained_variance_ratio_)
```

```python
# Graph the overview of the PCA explained variance

plt.plot(range(len(pca.explained_variance_ratio_)), pca.explained_variance_ratio_)
plt.show()
```

```python
# Graph the first 10 PCA components

plt.plot(range(10), pca.explained_variance_ratio_[0:10])
plt.show()
```

```python
pca = PCA(n_components=0.90)
pca.fit(crypto_df_prep_scale)
print(np.sum(pca.explained_variance_ratio_))
print(pca.explained_variance_ratio_)
print(len(pca.explained_variance_ratio_))
```

```python
# Create a dataframe with 11 PCA components to keep 90% of the explained variance

crypto_df_prep_scale_pca = pca.transform(crypto_df_prep_scale)
crypto_df_prep_scale_pca = pd.DataFrame(crypto_df_prep_scale_pca, columns = ["PCA"+ str(x) for x in range(1, len(pca.explained_variance_ratio_) + 1)])
crypto_df_prep_scale_pca.head()
```

```python
# Create the TSNE model

from sklearn.manifold import TSNE

tsne = TSNE(learning_rate=35)
tsne_features = tsne.fit_transform(crypto_df_prep_scale_pca)
tsne_features.shape
```

```python
# Visualize the TSNE clusters

plt.scatter(tsne_features[:,0], tsne_features[:,1])
plt.show()
```

```python
from sklearn.cluster import KMeans

# Find the best value for "k"

inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values

for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(crypto_df_prep_scale_pca)
    inertia.append(km.inertia_)

# Create the Elbow Curve

elbow_data = {"k": k, "inertia": inertia}
elbow_df = pd.DataFrame(elbow_data)

plt.plot(elbow_df['k'], elbow_df['inertia'])
plt.xticks(list(range(1, 11)))
plt.title('Elbow Curve')
plt.xlabel('# of Clusters')
plt.ylabel('Inertia')
plt.show()
```

k=8

```python
# Predict clusters with k=8

# Initialize the k-means model

model = KMeans(n_clusters=8, random_state=0)

# Fit the model

model.fit(crypto_df_prep_scale_pca)

# Predict the clusters

predictions = model.predict(crypto_df_prep_scale_pca)

# Add the predicted class column to the dataframe

crypto_df_prep_scale_pca["class"] = model.labels_
crypto_df_prep_scale_pca.head()
```

```python
# Visualize the clusters

plt.scatter(tsne_features[:,0], tsne_features[:,1], c=crypto_df_prep_scale_pca["class"])
plt.show()
```
